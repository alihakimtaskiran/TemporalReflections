```python
transcendence=[]
for i in range(15):
    transcendence.append(glimy.geo.Rectangle((500,0),(600,999), 0,3, time=(10*i,12*i)))
print(transcendence)
```

```python
entity=glimy.Continuum((1000,1000),5e-8)
entity.add(transcendence)
entity.add(glimy.DotSource((450,10),(0,100000),1,500e12))
entity.add(glimy.Boundaries(2,"ABC"))

entity.build(1,1)#replace 1,1 with 1,0 if you can't access GPU
entity.view_structure()
```

```python
entity.Render(101,"cupy")#replace "cupy" with "numpy" if you can't access GPU
entity.view_field()
entity.view_structure()
```
